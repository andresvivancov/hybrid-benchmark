```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('windowsvslatency5.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[1]))
        y.append(int(row[4]))

plt.plot(x,y, label='Flink')
plt.xlabel('Windows Size (ms)')
plt.ylabel('Latency (ms)')
plt.title('Windows Size vs Latency \non Flink')
plt.legend()
plt.show()
```
